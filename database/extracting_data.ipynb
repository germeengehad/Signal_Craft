```python
import yfinance as yf
import pandas as pd
import psycopg2
from sqlalchemy import create_engine

# 1. Download S&P500 data
data = yf.download("^GSPC", start="2010-01-01")
data.reset_index(inplace=True)  # make 'Date' a normal column

# 2. Connect to PostgreSQL
engine = create_engine("postgresql://postgres:YOUR_PASSWORD@localhost:5432/market_data")

# 3. Save data to SQL
data.to_sql("sp500_daily", engine, if_exists="replace", index=False)

print("Data saved to PostgreSQL successfully!")
```
